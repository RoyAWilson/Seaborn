```python
## Import Seaborn and load sp,me datasets data set
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
planets = sns.load_dataset('planets')
```

```python
tips
```

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue = 'day', size= 'size', palette='YlGnBu').set_title('Tip over Total Bill, Size and Day')
plt.savefig('../Figures/1_simpleScatter_2.png')
```

```python
## Histogram
# arg kde to show line as well as hist.
sns.histplot(tips['tip'], kde=True, bins = 15).set_title('Distribution of Tips')
plt.savefig('../Figures/2_simpleHist_1.png')
```

## With Displot for same plot

```python
sns.displot(data=tips['tip'], kde= True, bins = 15)
plt.savefig('../Figures/2_simpleHist_2.png')
```

# Bar Plot

```python
sns.barplot(x='sex', y='tip', data=tips, palette='YlGnBu').set_title('Tips by Gender')
plt.savefig('../Figures/3_simpleBarPlot_1.png')
```

# Box Plot

```python
sns.boxplot(data=tips, x='day', y='tip', hue='sex', palette='YlGnBu').set_title('Tips by Gender and Day')
plt.savefig('../Figures/4_simpleBoxPlot_1.png')
```

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', palette='YlGnBu')
plt.savefig('../Figures/4_simpleBoxPlot_2.png')
```

# Strip Plot

```python
sns.stripplot(data=tips, x='day', y='tip', hue='sex', palette='YlGnBu', dodge=True).set_title('Tips by Gender and Day')
## set dodge=True to seperate the genders on the plot
plt.savefig('../Figures/5_simpleStripPlot_1.png')
```

# Joint Plots

```python
sns.jointplot(data=tips, x='tip', y='total_bill', hue='sex', palette='YlGnBu')
plt.savefig('../Figures/5_simplejointPlot_1.png')
```

```python
sns.jointplot(data=tips, x='tip', y='total_bill')
plt.savefig('../Figures/5_simplejointPlot_01.png')
```

```python
sns.jointplot(data=tips, x='tip', y='total_bill', kind='reg')
plt.savefig('../Figures/5_simplejointPlot_02.png')
```

```python
sns.jointplot(data=tips, x='tip', y='total_bill', kind='kde', fill = True, cmap='YlGnBu')
plt.savefig('../Figures/5_simplejointPlot_03.png')
# Preferred the graph without the colourmap
```

```python
sns.jointplot(data=tips, x='tip', y='total_bill', kind='hex', cmap='YlGnBu')
plt.savefig('../Figures/5_simplejointPlot_04.png')
```

# Pair Plots

```python
sns.pairplot(data=titanic.select_dtypes(['number']), hue='pclass')
plt.savefig('../Figures/5_simplePairPlotPlot_01.png')
```

# Corelations

```python
titanic.corr()
```

```python
sns.heatmap(titanic.corr(), annot=True, cmap='YlGnBu').set_title('Correlation Heatmap')
plt.savefig('../Figures/6_corr_heatmap_01.png')
```

```python
sns.heatmap(titanic.corr(), annot=True, cmap='coolwarm').set_title('Correlation Heat Map')
plt.savefig('../Figures/6_corr_heatmap_02.png')
```

```python
sns.heatmap(titanic.corr(), annot=True, cmap='icefire').set_title('Correlation Heatmap')
plt.savefig('../Figures/6_corr_heatmap_03.png')
```

# Cluster Map

```python
sns.clustermap(data=iris.drop('species', axis=1), cmap='YlGnBu')
plt.savefig('../Figures/6_clustermap_01.png')
```
